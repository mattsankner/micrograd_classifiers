## In this notebook, we make a similar ```binary classifier``` to the one we made at the end of the @micrograd repository, but we make some changes to make the task more challenging:

-  Let's choose a dataset with more complexity: the ```circles``` dataset from ```scikit-learn```
-  We use a lower learning rate so jumps aren't too significant
-  We will also incorporate dropout layers to prevent overfitting ($0.2$)


````python
import random
import numpy as np
import matplotlib.pyplot as plot
%matplotlib inline
from micrograd.engine import Value
from micrograd.mlp import Neuron, Layer, MLP

np.random.seed(1337)
random.seed(1337)
from sklearn.datasets import make_circles

# Create dataset using 'make_circles' from sklearn
X, y = make_circles(n_samples=100, factor=0.5, noise=0.1)  # Add Gaussian noise to the data

# Transform labels to be -1 to 1 instead of 0 to 1 (default)
y = y * 2 - 1

# Plot dataset in 2D
plot.figure(figsize=(5, 5))
plot.scatter(X[:, 0], X[:, 1], c=y, s=20, cmap='jet')

# Define a deeper MLP model
class MLPComplex(MLP):
    def __init__(self, nin, nouts):
        super().__init__(nin, nouts)
        self.dropout_rate = 0.2  # Adjusted dropout rate
        print(f"Initialized MLPComplex with {nin} inputs and {nouts} neurons in each layer")
    
    def forward(self, x):
        out = x
        for layer in self.layers:
            out = layer(out)
            if layer != self.layers[-1]:  # Don't apply ReLU activation and dropout to the output (last) layer
                out = [o.relu() for o in out]  # Apply ReLU activation to each neuron's output
                out = self.apply_dropout(out)  # Apply dropout to the activations
        return out
    
    def apply_dropout(self, activations):
        if self.dropout_rate > 0:  # Dropout only applied if the dropout rate > 0
            for i in range(len(activations)):
                if random.random() < self.dropout_rate:  # If random number < dropout rate, drop the neuron
                    activations[i] = Value(0.0)  # Sets activation to 0 (drops neuron)
            print(f"Applied dropout with rate {self.dropout_rate}, {sum(1 for a in activations if a.data == 0)} neurons dropped")
        return activations

model = MLPComplex(2, [16, 16, 1])  # Same model depth
print("Parameter count:", len(model.parameters()))

def loss(batch_size=None):
    if batch_size is None:
        Xb, yb = X, y
    else:
        ri = np.random.permutation(X.shape[0])[:batch_size]
        Xb, yb = X[ri], y[ri]

    # Convert each input feature into Value object
    inputs = [list(map(Value, xrow)) for xrow in Xb]
    # Apply the model to each input to get the predicted scores
    scores = list(map(model, inputs))

    # Forward pass
    # Calculate hinge loss for each prediction (0, 1-y*score) = (1+-y*score).relu()
    losses = [(1 + -yi * scorei).relu() for yi, scorei in zip(yb, scores)]
    data_loss = sum(losses) * (1.0 / len(losses))  # Average hinge loss

    # Regularization term
    alpha = 1e-4  # L2 regularization term
    reg_loss = alpha * sum((p * p for p in model.parameters()))
    total_loss = data_loss + reg_loss  # Sum of data loss and regularization loss

    # Compare sign of true labels with sign of predicted scores for accuracy
    accuracy = [(yi > 0) == (scorei.data > 0) for yi, scorei in zip(yb, scores)]
    print(f"Data loss: {data_loss.data}, Reg loss: {reg_loss.data}, Total loss: {total_loss.data}")

    return total_loss, sum(accuracy) / len(accuracy)

total_loss, acc = loss()
print("Total loss:", total_loss.data, ", Accuracy:", acc * 100, "%")

# Optimization (learning rate linearly from 1.0 to 0.1 over 100 iterations)
for k in range(100):
    total_loss, acc = loss()  # Total loss and accuracy calculated using loss func
    model.zero_grad()  # Reset gradients
    total_loss.backward()  # Backpropagation performed to compute gradients
    learning_rate = 0.1 - 0.001 * k  # Adjusted learning rate decay
    for p in model.parameters():  # Gradient descent
        p.data -= learning_rate * p.grad
    
    if k % 1 == 0:
        print(f"Step {k + 1} loss {total_loss.data}, accuracy {acc * 100}%")

# Visualize decision boundary
h = 0.25  # Step size for mesh grid
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1  # Grid spans from xmin to xmax
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1  # And ymin to ymax with steps of h

# Generate two 2D arrays, 'xx' and 'yy' for grid coordinates
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),  # Array from x_min to x_max, step size h
                     np.arange(y_min, y_max, h))  # Array from y_min to y_max
Xmesh = np.c_[xx.ravel(), yy.ravel()]  # Flatten into 1D array, combined column-wise to create 2D array
# Convert grid points to Value objects
inputs = [list(map(Value, xrow)) for xrow in Xmesh]
# Predictions are made
scores = list(map(model, inputs))
# Convert predictions to boolean values
Z = np.array([s.data > 0 for s in scores])
# Reshape back into a 2D array that matches shape of grid
Z = Z.reshape(xx.shape)

print(f"Decision boundary shape: {Z.shape}")
print(f"Number of positive predictions: {np.sum(Z)}")
print(f"Number of negative predictions: {np.sum(~Z)}")

fig = plot.figure()
plot.contourf(xx, yy, Z, cmap=plot.cm.Spectral, alpha=0.8)
plot.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=plot.cm.Spectral)
plot.xlim(xx.min(), xx.max())
plot.ylim(yy.min(), yy.max())

print(f"\nFinal Results:")
final_loss, final_acc = loss()
print(f"Total loss: {final_loss.data:.4f}")
print(f"Final accuracy: {final_acc * 100:.2f}%")
````

## Since that worked quite well, let's try a much deeper model!

````python
model = MLPComplex(2, [32, 32, 16, 1])  # Original model depth: model = MLP(2, [16, 16, 1])
print("Parameter count:", len(model.parameters()))

# Optimization (learning rate linearly from 1.0 to 0.1 over 100 iterations)
for k in range(100):
    total_loss, acc = loss()  # Total loss and accuracy calculated using loss func
    model.zero_grad()  # Reset gradients
    total_loss.backward()  # Backpropagation performed to compute gradients
    learning_rate = 0.1 - 0.001 * k  # Adjusted learning rate decay
    for p in model.parameters():  # Gradient descent
        p.data -= learning_rate * p.grad
    
    if k % 1 == 0:
        print(f"Step {k + 1} loss {total_loss.data}, accuracy {acc * 100}%")
````
